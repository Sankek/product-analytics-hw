## Цель: Рост DAU

DAU (Daily Active Users) – это ключевая метрика, которая измеряет количество уникальных активных пользователей в день. Рост DAU может быть достигнут за счёт увеличения количества пользователей и/или их удержания. Для этого проанализируем ключевые факторы, влияющие на DAU, используя подход на основе фреймворка AARRR, чтобы структурировать метрики и привязать их к жизненному циклу пользователя.

### Подход: Фреймворк AARRR 
- Acquisition (Привлечение): Метрики, показывающие, как новые пользователи попадают в Telegram.
- Activation (Активация): Метрики, которые измеряют, активировался ли пользователь, то есть совершил ли он действия, которые приводят к использованию сервиса.
- Retention (Удержание): Метрики, показывающие, остаются ли пользователи активными после первоначального использования.
- Revenue (Выручка): Хотя Telegram не ориентируется на прямую монетизацию пользователей, мы можем включить сюда метрики, связанные с подписками на Premium.
- Referral (Рекомендации): Метрики, которые показывают, привлекают ли текущие пользователи новых.

#### Разработка системы метрик и дашборда

Сформируем систему метрик для каждого из этапов AARRR, которые повлияют на рост DAU:

#### 1. Acquisition (Привлечение)
Эти метрики помогают понять, как пользователи впервые попадают в приложение и что способствует росту новой аудитории:

- Количество новых регистраций (New Registrations) – общее число новых пользователей в день.
- Источники трафика (Traffic Source) – доля пользователей, привлечённых с разных источников (например, рекламные каналы, органический трафик, социальные сети).
- Конверсия посетителей в регистрации (Visitor-to-Signup Conversion) – процент пользователей, которые, установив приложение, зарегистрировались.

#### 2. Activation (Активация)
Метрики активации показывают, вовлекаются ли пользователи в приложение сразу после регистрации:

- Доля пользователей, отправивших первое сообщение (First Message Rate) – доля новых пользователей, отправивших сообщение в - день регистрации.
- Время до первого взаимодействия (Time to First Action) – среднее время от регистрации до первого действия (например, отправки сообщения).
- Доля пользователей, добавивших контакт или вступивших в группу (Contacts/Group Join Rate) – процент пользователей, которые добавили первый контакт или вступили в группу/канал в первые дни.
#### 3. Retention (Удержание)
Метрики удержания – одни из самых важных для DAU, так как они показывают, сколько пользователей возвращаются:

- Удержание по когортам (Cohort Retention Rate) – показатель удержания пользователей, пришедших в разные временные периоды, на следующий день, через 7 и 30 дней.
- Удержание на день X (Retention Day X) – доля пользователей, возвращающихся на определённый день после первой активации (например, D1, D7, D30).
- Среднее время в приложении (Average Time Spent) – среднее время, которое пользователь тратит в приложении за день.
DAU/MAU (отношение дневной к месячной активности) – показатель вовлечённости: чем ближе к 1, тем чаще пользователи возвращаются в приложение ежедневно.
#### 4. Revenue (Выручка)
Поскольку Telegram частично монетизируется через подписку, важно отслеживать следующие метрики:

- Количество новых и активных подписок Premium (New and Active Premium Subscriptions) – общее количество и динамика новых подписок на Telegram Premium.
- ARPU (Средняя выручка на пользователя) – средняя выручка, генерируемая активными пользователями, особенно важна для Telegram Premium.
- Процент пользователей, переходящих на Premium (Premium Conversion Rate) – доля пользователей, которые конвертируются из бесплатного использования в платное.
#### 5. Referral (Рекомендации)
Эти метрики показывают, насколько пользователи приводят новых пользователей в Telegram, что важно для органического роста:

- Процент пользователей, пригласивших хотя бы одного друга (User Invite Rate) – процент пользователей, которые используют функцию приглашения.
- Вовлечение через реферальные ссылки (Referral Engagement) – процент новых пользователей, пришедших по реферальной ссылке.

# Задание 2.
_В поисковую выдачу билетов сайта aviasales была разработана и внедрена скоринговая модель, ранжирующая билеты. Ее цель - помочь пользователю выбрать лучший авиабилет самым быстрым способом, подняв его в топ выдачи._

_Пожалуйста, внимательно изучи поисковую выдачу сайта и предложи:_

_Какими метриками бы ты отслеживал качество работы модели?_


Для анализа качества скоринговой модели на Aviasales, проведем детальный разбор ключевых метрик

## Анализ целей и постановка задачи
Цель модели — повысить удобство поиска, отображая на первых позициях билеты, которые с наибольшей вероятностью удовлетворят пользователя. Задача анализа качества — понять, насколько эффективно модель достигает этой цели, и оценить пользовательский опыт в поисковой выдаче.

## Определение ключевых метрик
Метрики можно разделить на три группы:

- Метрики ранжирования — оценивают, насколько релевантны топовые результаты поисковой выдачи.
- Пользовательские метрики — отслеживают взаимодействие пользователей с результатами поиска.
- Поведенческие метрики — анализируют общее поведение пользователей и уровень их удовлетворенности.

### Метрики ранжирования. 
Эти метрики позволяют понять, насколько высококачественные результаты модель поднимает в топ:

####  Mean Reciprocal Rank (MRR)
Средний обратный ранг первого релевантного билета, где релевантным считается билет, по которому совершено действие (например, клик или бронь). Чем выше MRR, тем лучше модель поднимает релевантные результаты в топ.

- Формула: $\text{MRR}=\frac{1}{|Q|}\sum_{i=1}^{|Q|}\frac{1}{\text{rank}_i}$, где ${\text{rank}_i}$ — позиция первого релевантного билета в результатах поиска для запроса $Q$

#### Normalized Discounted Cumulative Gain (NDCG)
Оценивает релевантность и порядок результатов выдачи. Эта метрика учитывает позицию релевантных билетов, поощряя те, которые находятся в топе.

- Формула: $\text{NDCG}_p=\frac{\text{DCG}_p}{\text{IDCG}_p}$, где $\text{DCG}_p$ — кумулятивный показатель релевантности на первых $p$ позициях, а $\text{IDCG}_p$ — максимальный $\text{DCG}$ возможный для идеального ранжирования.

#### Precision at K (P@K)
Точность в топ-K результатах. Показывает долю релевантных результатов среди первых $K$ позиций. Полезна для понимания, как модель справляется с первыми результатами, куда чаще всего смотрят пользователи.

### Пользовательские метрики

#### CTR (Click-Through Rate)
Соотношение числа кликов на результаты поиска к числу показов. Высокий CTR указывает на релевантность топовых результатов и привлекательность предложений.

- Формула: $\text{CTR}=\frac{\text{Число кликов}}{\text{Число показов}}$.

#### Conversion Rate (CR)
Доля пользователей, совершивших бронирование, к числу тех, кто взаимодействовал с результатами поиска. Это ключевая метрика для оценки того, насколько предлагаемые билеты соответствуют потребностям пользователя.

- Формула: $\text{CR}=\frac{\text{Число бронирований}}{\text{Число взаимодействий с поисковой выдачей}}$.

### Поведенческие метрики
Эти метрики оценивают общее удобство использования:

#### Среднее время на поисковой странице
Время, которое пользователи проводят на странице поиска. Если оно уменьшается при улучшении модели, значит, пользователи быстрее находят подходящий билет.

#### Процент возвратов к поиску
Процент пользователей, которые вернулись к поиску после просмотра одного или нескольких билетов. Высокий процент возвратов может указывать на то, что топовые результаты не всегда удовлетворяют запрос пользователя.


## Обоснование
Предложенные метрики помогут понять, насколько эффективно модель выполняет свою задачу. С точки зрения продукта, важно не только ранжировать билеты, но и обеспечивать удобство поиска, что отражается в таких метриках, как $\text{CTR}$ и $\text{CR}$.

Эти метрики позволят:

- Отслеживать релевантность результатов, выводимых на топовые позиции.
- Понять поведение пользователей при взаимодействии с поисковой выдачей.
- Анализировать эффективность улучшений модели и их влияние на качество поиска билетов.
